```python
# 할일을 리스트에 추가하는 함수
def add_task(task_list, task): 
    task_list.append(task) # 할일 추가

# 할일 목록을 보여주는 함수
def display_tasks(task_list):
    i = 1
    if not task_list:
        print("할일 목록이 비어 있습니다.")
    else:
        print("할일 목록: ")
        for task in task_list: # 할일 목록을 출력
            print(f'{i}. {task}')
            i += 1

# 할일 목록을 파일에 저장하는 함수
def save_tasks_to_file(task_list, filename):
    fd = open(filename ,"w", encoding="utf-8") # 파일 쓰기 모드로 열기
    for task in task_list:
        fd.write(task + '\n') # 할일을 저장하기(줄바꿈 포함)
    fd.close()
    
# 할일 목록을 파일에서 읽고 리스트에 추가하는 함수
def load_tasks_from_file(filename):
    task_list = []
    fd = open(filename, "r", encoding="utf-8") # 파일 읽기 모드로 열기
    fd_lines = fd.readlines() # 파일 내용 읽기
    for f_line in fd_lines:
        task_list.append(f_line.strip())
    return task_list

file_name = "task.txt"
task_list = load_tasks_from_file(file_name)

while True:
    print("\n할일 관리 프로그램")
    print("1. 할일 추가")
    print("2. 할일 목록 표시")
    print("3. 종료")
    
    choice = input("선택: ")
    
    
    if choice == '1':
        task = input("할일을 입력하세요: ")
        add_task(task_list, task)
        print("할일이 추가되었습니다.")
    elif choice == '2':
        display_tasks(task_list)
    elif choice == '3':
        save_tasks_to_file(task_list, 'task.txt')
        print("프로그램을 종료합니다.")
        break
    else:
        print("잘못된 선택입니다. 다시 시도하세요.")
```
